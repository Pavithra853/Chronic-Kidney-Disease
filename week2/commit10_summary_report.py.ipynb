```python
import pandas as pd
import os

os.makedirs("output", exist_ok=True)

df = pd.read_csv("C:/Users/Pavithra/OneDrive/Documents/Data Analytics Project 3/data/Chronic_Kidney_Disease.csv")

for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

numeric_df = df.select_dtypes(include="number")

correlation = numeric_df.corr()

high_corr = correlation[
    (correlation > 0.75) &
    (correlation < 1)
]

with open("output/week2_summary.txt","w") as file:

    file.write("Week 2 Exploratory Data Analysis Summary\n")
    file.write("="*50+"\n\n")

    file.write("Dataset Shape\n")
    file.write(str(df.shape))
    file.write("\n\n")

    file.write("Numerical Columns\n")
    file.write(str(numeric_df.columns.tolist()))
    file.write("\n\n")

    file.write("Highly Correlated Features\n")
    file.write(str(high_corr))
    file.write("\n")

print("Week 2 Summary Report Generated Successfully")
```
