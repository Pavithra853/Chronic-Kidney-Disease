```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

os.makedirs("output", exist_ok=True)

df = pd.read_csv("C:/Users/Pavithra/OneDrive/Documents/Data Analytics Project 3/data/Chronic_Kidney_Disease.csv")

numeric_columns=['age','bp','hemo','sc']

for col in numeric_columns:
    if col in df.columns:
        df[col]=pd.to_numeric(df[col],errors='coerce')

sns.set(style="whitegrid")

# Age vs Blood Pressure

if 'age' in df.columns and 'bp' in df.columns:

    plt.figure(figsize=(8,6))

    sns.scatterplot(data=df,
                    x='age',
                    y='bp')

    plt.title("Age vs Blood Pressure")

    plt.savefig("output/age_vs_bp.png")

    plt.close()

# Hemoglobin vs Serum Creatinine

if 'hemo' in df.columns and 'sc' in df.columns:

    plt.figure(figsize=(8,6))

    sns.scatterplot(data=df,
                    x='hemo',
                    y='sc')

    plt.title("Hemoglobin vs Serum Creatinine")

    plt.savefig("output/hemo_vs_sc.png")

    plt.close()

print("Comparison plots completed.")
```
