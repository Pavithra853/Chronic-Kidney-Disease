```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

os.makedirs("output", exist_ok=True)

df = pd.read_csv("C:/Users/Pavithra/OneDrive/Documents/Data Analytics Project 3/data/Chronic_Kidney_Disease.csv")

# Convert important biomarkers into numeric

columns = ['age','bp','sg','al','hemo']

for col in columns:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

sns.set(style="whitegrid")

for col in columns:

    if col in df.columns:

        plt.figure(figsize=(8,5))

        sns.histplot(df[col].dropna(),
                     kde=True,
                     bins=20)

        plt.title(f"{col} Distribution")

        plt.savefig(f"output/{col}_distribution.png")

        plt.close()

print("Biomarker distribution plots created successfully.")
```
