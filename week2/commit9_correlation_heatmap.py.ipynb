```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

os.makedirs("output", exist_ok=True)

df = pd.read_csv("C:/Users/Pavithra/OneDrive/Documents/Data Analytics Project 3/data/Chronic_Kidney_Disease.csv")

# Convert all possible numeric columns

for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

numeric_df = df.select_dtypes(include="number")

correlation = numeric_df.corr()

plt.figure(figsize=(14,10))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5,
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.tight_layout()

plt.show()      # Display heatmap on screen

plt.savefig("output/heatmap.png")  # Optional
plt.close()
print("\nHeatmap Generated Successfully")
```
